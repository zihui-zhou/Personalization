```python
TextFileReader = pd.read_csv('yelp_dataset/review.csv', chunksize=100000)

dfList = []
for df in TextFileReader:
    dfList.append(df)

df = pd.concat(dfList, sort=False).reset_index(drop=True)
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
df.head(5)
```

```python
df['text'] = df['text'].fillna("")
```

```python
df.loc[ df.isnull().any(axis=1) ]
```

```python
def convertStr(x):
    try:
        return re.sub('\s+', ' ', x)
    except:
        print(x)
        return x
df['text'] = df['text'].apply(lambda x: convertStr(x))
df.loc[2823429]['text']
```

```python
rr = df.loc[
    df['business_id'].isin(business_id['business_id'].values.tolist()) &\
    df['user_id'].isin(user_id['user_id'].values.tolist())
]
```

```python
import pandas as pd
t = pd.read_csv('yelp_dataset/review_cleaned.csv', index_col=0)
```

```python
lastComment = t.sort_values(by='date')\
               .groupby('user_id')\
               .agg({'date': 'last'})\
               .reset_index()

lastComment.head()
```

```python
print(lastComment.shape)
```

```python
t.set_index(['user_id', 'date'], inplace=True)
t.head()
```

```python
t['TrainTest'] = "1"
t.head()
```

```python
t['TrainTest'].loc[list(lastComment.itertuples(index=False, name=None))] = 0
```

```python
t.reset_index(inplace=True)
t['user_id'].nunique()
```
